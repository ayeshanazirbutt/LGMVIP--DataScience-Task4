```python
import seaborn as sns
import matplotlib.pyplot as plt

top_groups = df['gname'].value_counts().drop('Unknown').nlargest(10)
plt.figure(figsize=(12, 6))
sns.barplot(x=top_groups.index, y=top_groups.values, palette='viridis')

plt.xticks(rotation=45, ha='right', fontsize=12)
plt.xlabel('Terrorist Group', fontsize=14)
plt.ylabel('Number of Attacks', fontsize=14)
plt.title('Top 10 Terrorist Groups by Number of Attacks', fontsize=16)

plt.tight_layout()
plt.show()
```


<center><h2>6. COUNTRY WISE DEATHS, WOUNDED AND TOTAL(AFFECTED) ON ATTACKS</h2></center>

```python
df3=df['country_txt'].value_counts()[:15].to_frame()
df3.columns=['Attacks']
data=df.groupby('country_txt')['nwound'].sum().to_frame()
df3.merge(data,left_index=True,right_index=True,how='left').plot.bar(width=0.9,color=['red','pink'])
fig=plt.gcf()
fig.set_size_inches(18,6)
plt.title('Counntry-wise Attacks vs Wounded')
plt.xlabel('Country')
plt.ylabel('count')
plt.show()
```

```python
df3=df['country_txt'].value_counts()[:15].to_frame()
df3.columns=['Attacks']
data=df.groupby('country_txt')['nkill'].sum().to_frame()
df3.merge(data,left_index=True,right_index=True,how='left').plot.bar(width=0.9,color=['blue','yellow'])
fig=plt.gcf()
plt.title('Counntry-wise Attacks vs Killed')
fig.set_size_inches(18,6)
plt.xlabel('Country')
plt.ylabel('count')
plt.show()
```
